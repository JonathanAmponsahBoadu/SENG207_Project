```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('red_wine.csv')
```

```python
dataset.head(10)
```

```python
dataset.tail(20)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
#creating an instance of the dataset to fill the null values with mean for each column
dataset_filled = dataset.copy()
```

```python
#finding mean for each column in the instance that has a null value
mean_fa = dataset_filled['fixed acidity'].mean()
mean_va = dataset_filled['volatile acidity'].mean()
mean_ch = dataset_filled['chlorides'].mean()
mean_fsd = dataset_filled['free sulfur dioxide'].mean()
mean_ds = dataset_filled['density'].mean()
mean_qlt = dataset_filled['quality'].mean()
```

```python
#filling the new dataset with the mean of each column that has a null value respectively
dataset_filled['fixed acidity'] = dataset_filled['fixed acidity'].fillna(mean_fa)
dataset_filled['volatile acidity'] = dataset_filled['volatile acidity'].fillna(mean_va)
dataset_filled['chlorides'] = dataset_filled['chlorides'].fillna(mean_ch)
dataset_filled['free sulfur dioxide'] = dataset_filled['free sulfur dioxide'].fillna(mean_fsd)
dataset_filled['density'] = dataset_filled['density'].fillna(mean_ds)
dataset_filled['quality'] = dataset_filled['quality'].fillna(mean_qlt)
```

```python
dataset_filled.corr()
```

```python
   correlation_matrix = dataset_filled.corr()
   plt.figure(figsize=(10, 8))
   plt.imshow(correlation_matrix, cmap="coolwarm", interpolation = "none" , aspect = "auto")
   plt.colorbar()
   plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation= "vertical")
   plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
   plt.title("Correlation Matrix")
```

```python
dataset_filled.hist(figsize = (12,10))
plt.suptitle("Histograms of properties")
```

```python
dataset_filledReview = dataset.copy()
```

```python
dataset_filledReview['review'] = dataset_filledReview['quality'].apply(lambda x: 1 if x >= 7 else 0)
```

```python
review_counts = dataset_filledReview['review'].value_counts()
labels = ['Not Good (0)', 'Good (1)']
sizes = review_counts.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Proportion of Wine Types Based on Reviews')
```

```python
quality_counts = dataset_filled['quality'].value_counts().sort_index()
colors = plt.cm.viridis(np.linspace(0, 1, len(quality_counts)))
plt.bar(quality_counts.index, quality_counts.values, color=colors,width = 0.3)
plt.xlabel('Quality')
plt.ylabel('Number of Wines')
plt.title('Number of Wines for Each Quality')
plt.xticks(quality_counts.index) 
```

```python
review_counts = dataset_filledReview['review'].value_counts().sort_index()
colors = plt.cm.viridis(np.linspace(0, 1, len(review_counts)))
plt.bar(review_counts.index, review_counts.values, color=colors)
plt.xlabel('Review')
plt.ylabel('Number of Wines')
plt.title('Number of Wines for Each Review')
plt.xticks(review_counts.index, ['Not Good (0)', 'Good (1)'])
```

### Report

##### Correlation table and correlation matrix
- From the correlation table and correlation matrix it was observed that alcohol, citric acid and sulphates share a positive correlation with the wine quality
- Fixed acidity also shows a weak positive correlation with the wine quality
- However, alcohol shows the highest correlation
- This indicates that alcohol content is essential in making wine of higher quality
- Hence modereately high amounts of alcohol is crucial in making wine of high quality
- Total sulphur dioxide and volatile acidity also shows moderate negative correlations with the quality.
- Volatile acidity shows higher negative correlation
- This shows that less amounts of valatile acidity are required for wine of higher quality
- The other properties do not show significant correlations with the quality of the wine

##### Histograms
- The histogram indicates the number of occurrences of each property of the wine in the dataset
- Fixed acidity has the highest value occurrence being 7.5 and occurring about 400-500 times
- Volatile acidity also has the highest value occurrence being about 6 and occurring about 450 times
- Citric acid and residual sugar have their highest values occurrence being being close to zero and ocuuring about 400 and 1000 times respectively
- Negative skewness is exhibited by pH and sulphates indicating that most of the similar values occur for the small values and decrease proportionally
- Free sulphur dioxide and total sulphur dioxide also exhibit negative skewness and have the highest occurring value being close to 0
- Chlorides,density,pH,sulphates and alcohol shows no particular skewness and have highest occurring values being close to 0
- The pH is highly central around the range of 2-4 indicating that all wine combinations are acidic.
- Quality has no particular skewness.The highest occurring quality is 5 which is a poor quality
- Quality between 7 and 8 constitutes a small part of the table since the number of occurrence is small

##### Pie chart
- The quality was grouped into two: Good(1) for quality greater or equal to 7 and Bad(0) for quality less than 7
- The pie chart indicates that wine of good quality constitutes 13.6% of the table and wine of bad quality constitutes 86.4%
- This indicates that most of the combinations are bad and only a few are good to produce red wine of high quality

##### Bar Chart
- The bar chart for the red wine quality indicates that quality rating of 5 occurs about 680 times
- That of 6 occurs about 630 times
- Most wine combinations have either of these two ratings

- Less  ratings such as 3,4 and 5.63 do not occur more often in the table and hence only a few combinations have very bad quality

##### Conclusion
- Red wine of high quality have high alcohol contents , moderate citric acid contents and sulphates and of volatile acidity
